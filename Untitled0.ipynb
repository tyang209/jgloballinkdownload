```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import datetime

def checkExists(driver,xpath):
    try:
        driver.find_element_by_xpath(xpath)
    except NoSuchElementException:
        return False
    return True


array=[]
driver = webdriver.PhantomJS('phantomjs.exe')
driver.set_window_size(1920, 1080)

driver.get("http://jglobal.jst.go.jp/search.php#%7B%22keyword%22%3A%22%E4%BA%AC%E9%83%BD%E5%A4%A7%E5%AD%A6%22%2C%22synonym%22%3Anull%2C%22category%22%3A1%2C%22order%22%3A%22score%22%2C%22limit%22%3A100%2C%22page%22%3A2%2C%22words%22%3A%5B%5D%7D")
driver.implicitly_wait(30)
button_xpath = "//a[contains(@class,'nextPageBtn')]"
table_xpath = "//table[contains(@class,'type1')]"

with open('profileLinksCollection.txt','wb') as f:
    while checkExists(driver, button_xpath):
        table_contents = driver.find_element_by_xpath(table_xpath)
        soup = BeautifulSoup(table_contents.get_attribute('innerHTML'))
        for tr in soup.findAll('tr'):
            f.write(tr.a['href']+'\n')       
        nextButton = driver.find_element_by_xpath(button_xpath)
        nextButton.click()

# table_contents = driver.find_elements_by_xpath("//table[contains(@class,'type1')]")
# array.append(table_contents)
# nextButton = driver.find_element_by_xpath("//a[contains(@class,'nextPageBtn')]")
# nextButton.click()
# array.append(driver.find_elements_by_xpath("//table[contains(@class,'type1')]"))


    

# for line in 

# html = driver.page_source
# driver.close()
# soup = BeautifulSoup(html)
```
